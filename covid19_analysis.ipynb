```python
df1=dataset.drop(["Lat","Long"],axis=1,inplace=True)
dataset.head()
```

```python
corona_dataset_aggregated=dataset.groupby("Country/Region").sum(numeric_only=True)
```

```python
countries =list(corona_dataset_aggregated.index)
max_infection_rates=[]
for i in countries:
  max_infection_rates.append(corona_dataset_aggregated.loc[i].diff().max())
corona_dataset_aggregated["Max infection rates"]=max_infection_rates
```

```python
corona_dataset_aggregated
```

```python
corona_data=pd.DataFrame(corona_dataset_aggregated["Max infection rates"])
corona_data
```
